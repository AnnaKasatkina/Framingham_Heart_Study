# Разведочный анализ данных

```python
import pandas as pd
```

## Загрузка данных и первый взгляд

```python
df = pd.read_csv('hf://datasets/Nnaodeh/Framingham_Stroke_Datasets/framingham Heart Study Dataset.csv')
df.head()
```

```python
print(df.shape)
df.describe()
```

## Выявление аномалий и выбросов

```python
numeric_cols = [
    'age', 'cigsPerDay', 'BPMeds', 'totChol',
    'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose'
]

df_clean = df.copy()
outlier_mask = pd.Series(False, index=df_clean.index)

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    low= q1 - 2 * iqr
    high = q3 + 2 * iqr

    col_outliers = (df_clean[col] < low) | (df_clean[col] > high)
    outlier_mask = outlier_mask | col_outliers

df_clean = df_clean[~outlier_mask]
print(df_clean.shape)
df_clean.describe()
```

## Пропуски и их заполнение

```python
df_clean.isnull().sum()
```

Тактики заполнения пропуска у признаков:
- ``education``: мода
- ``cigsPerDay``: 0 для некурящих, среднее для курящих
- ``BPMeds``: 0 при нет гипертонии, медиана при гипертонии
- ``totChol``, ``BMI``, ``heartRate``, ``glucose``: медиана

```python
# education
mode_edu = df_clean['education'].mode()[0]
df_clean['education'] = df_clean['education'].fillna(mode_edu)

# cigsPerDay
mean_cigs = df_clean.loc[df_clean['currentSmoker'] == 1, 'cigsPerDay'].mean()
df_clean.loc[(df_clean['currentSmoker'] == 0) & df_clean['cigsPerDay'].isna(), 'cigsPerDay'] = 0
df_clean.loc[(df_clean['currentSmoker'] == 1) & df_clean['cigsPerDay'].isna(), 'cigsPerDay'] = mean_cigs

# BPMeds
median_bp = df_clean.loc[df_clean['prevalentHyp'] == 1, 'BPMeds'].mean()
df_clean.loc[(df_clean['prevalentHyp'] == 0) & df_clean['BPMeds'].isna(), 'BPMeds'] = 0
df_clean.loc[(df_clean['prevalentHyp'] == 1) & df_clean['BPMeds'].isna(), 'BPMeds'] = median_bp

# totChol, BMI, heartRate, glucose
for col in ['totChol', 'BMI', 'heartRate', 'glucose']:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)
```
